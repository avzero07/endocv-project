```python
# Import Statements

%matplotlib inline

import os
import json
import numpy as np
from matplotlib import image
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
# Utility Functions

# Function to Annotate Image Based on Bounding Box Info
def overlayBBox(img,imgFileName,annot_dict):
    
    # Define Figure and Subplot
    fig, ax = plt.subplots(1)
    ax.imshow(img)
    
    # Since Some Images Have Multiple Annotations
    for key in annot_dict[imgFileName[:-4]]:
        left = int(annot_dict[imgFileName[:-4]][key]['left'])
        top = int(annot_dict[imgFileName[:-4]][key]['top'])
        right = int(annot_dict[imgFileName[:-4]][key]['right'])
        bottom = int(annot_dict[imgFileName[:-4]][key]['bottom'])
        
        # Define Rectangle
        rect = patches.Rectangle((left,top),bottom-top,right-left,linewidth=1,edgecolor='r',facecolor='none')
        # Add Patch
        ax.add_patch(rect)
        plt.text(left,top-5,annot_dict[imgFileName[:-4]][key]['class'],color='red')
    plt.show()
```

```python
# Working Directory
pathDir = 'C:/Users/aksha/Documents/Github Projects/endocv-project/Dataset/'

# Load Annotations
with open(pathDir+'bbox-json-resized.txt','r') as file:
    annot_dict = json.load(file)
    
with open(pathDir+'bbox-json.txt','r') as file:
    annot_orig = json.load(file)
```

```python
# Load an Image and Annotate
imgFileName = 'EDD2020_AJ0062.jpg'
img = image.imread(pathDir+'originalImages-resized/'+imgFileName)
overlayBBox(img,imgFileName,annot_dict)
```
